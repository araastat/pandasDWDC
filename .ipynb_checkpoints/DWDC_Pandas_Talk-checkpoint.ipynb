# Snakes eating data

Pandas is a Python data analysis package developed by Wes McKinney in 2008

+ Currently at version 0.16
+ Well-documented
+ http://pandas.pydata.org

## Why use pandas?

+ Created right kinds of data structures

    + Mixed type container (**DataFrame**)!!
    + Indexed data structures (refer by name or position)
    
+ Convenient

    + Nice wrappers around most functionality
    + Takes advantage of chained operations


## Why not use pandas?

+ Convenient!!

    + Not necessarily most efficient for production
    + Loads all data into memory
    + Requires long time to compile, for use in `virtualenv`

Nice blog by Ben Bengfort on [reading CSV in Python efficiently](https://districtdatalabs.silvrback.com/simple-csv-data-wrangling-with-python)

# Re-naming variables

```python
diamonds = pd.read_csv('diamonds.csv')
diamonds.rename(columns = {'x':'length','y':'width','z':'depth'}, inplace=True)
diamonds.columns.values[4] = 'depthperc'
diamonds.head()
```

# Creating new columns

```python
diamonds['cubic'] = diamonds['length']*diamonds['width']*diamonds['depth']
# diamonds['cubic'] = diamonds[['length','width','depth']].product(axis=1)
diamonds['total'] = diamonds.length+diamonds.width+diamonds.depth
# diamonds['total'] = diamonds[['length','width','depth']].sum(axis=1)
#
# diamonds.assign(cubic = diamonds[['length','width','depth']].product(axis=1),
#    total=diamonds[['length','width','depth']].sum(axis=1))

diamonds.head()
```

## Split-apply-combine

```python
diamonds.groupby('color')['price'].max()
```

```python
newdat=diamonds[['cut','clarity','depthperc','table','price','length','width','depth','cubic']]
newdat.groupby(['cut','clarity']).aggregate(np.mean)
#newdat.groupby(['cut','clarity'], as_index=False).aggregate(np.mean)
```

```python
cars = pd.read_csv('mtcars.csv')
grouped=cars.groupby(['cyl','gear','carb'])
stats = ['count','mean','median']
grouped.agg(stats)[['mpg','disp']]
```

```python
diamonds['carat2'] = np.round(diamonds.carat/0.25)*0.25
Summary = (
    diamonds[['cut','color','clarity','carat2','depthperc',
              'table','price','length','width','depth','cubic']]
    .groupby(['cut','color','clarity','carat2'],as_index=False)
    .aggregate(np.mean) # Does means of the rest
    )
Summary.head()
```

# Pivot tables

```python
PT = pd.pivot_table(diamonds[['color','clarity','price']],
                    index='color',
                    columns = 'clarity',
                    aggfunc = np.mean)
PT
```

# Vlookups

```python
Summary.rename(columns = {'price':'avgprice'}, inplace=True)
# Summary.head()
VL = pd.merge(diamonds, Summary[['cut','color','clarity','carat2','avgprice']],
              on = ['cut','color','clarity','carat2'])
VL.head()
```

# Conditioning

```python
diamonds[diamonds.cut=='Ideal'].price.describe()
```

```python
diamonds[diamonds.price > 1000].cut.value_counts()
```

```python
diamonds['Size'] = 'Small'
diamonds.loc[diamonds['carat']>= 0.5, 'Size'] = 'Medium'
diamonds.loc[diamonds['carat'] > 1, 'Size'] = 'Large'
diamonds['Size'] = pd.Categorical(diamonds['Size'],
    categories = ['Small','Medium','Large'])
diamonds.head()
```

# Charts

```python
%matplotlib inline
x = diamonds['Size'].value_counts(sort=False)
x.plot(kind='bar');
```

```python
diamonds.boxplot(column = 'price', by='cut');
```

```python
diamonds.plot(kind='scatter', x='carat',y='price');
```

```python
groups = diamonds.groupby('clarity')
fig, ax = plt.subplots();
for name, group in groups:
    ax.plot(group.carat, group.price, marker='o', linestyle='', label=name)
ax.legend(numpoints=1)
plt.xlabel('Carat');
plt.ylabel('Price');
```

```python
import seaborn as sns
sns.set(style='white', palette='muted')
sns.distplot(diamonds['price'], kde=False);
```
